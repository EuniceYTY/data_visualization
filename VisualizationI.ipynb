## Bar Charts

First, we will make some simple bar charts with matplotlib.  The examples come from [here](https://pythonspot.com/matplotlib-bar-chart/)

```python
%matplotlib inline
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np

 
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]
 
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')
 
plt.show()
```

```python
# data to plot
n_groups = 4
means_frank = (90, 55, 40, 65)
means_guido = (85, 62, 54, 20)
 
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
 
rects1 = plt.bar(index, means_frank, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Frank')
 
rects2 = plt.bar(index + bar_width, means_guido, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Guido')
 
plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width/2, ('A', 'B', 'C', 'D'))
plt.legend()
 
plt.tight_layout()
plt.show()
```

## Line Graphs

Next, some simple line graphs.  The examples come from [here](http://www.ds100.org/sp18/assets/lectures/lec07/07-intro-matplotlib.html)

```python
# Plot some random data uniformly distributed between 0 and 1
x = np.random.rand(100)
plt.plot(x)
```

```python
# Plot the function y = sin(x)

x = np.linspace(0, 2*np.pi, 300) # linspace produces a vector of 300 numbers ranging from 0 to 2pi
y = np.sin(x)
plt.plot(x, y);  # note the ';' at the end of the line, it suppresses the Out[N] block.
```

The most frequently used function is simply called plot, here is how you can make a simple plot of $sin(x)$ and  $sin(x^2)$ for x[0,2]  with labels and a grid (we use the semicolon in the last line to suppress the display of some information that is unnecessary right now):

```python
y2 = np.sin(x**2)
plt.plot(x, y, label=r'$\sin(x)$')
plt.plot(x, y2, label=r'$\sin(x^2)$')
plt.title('Some functions')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend();
```

## Plotting 2D Arrays (Images)

An image is just a 2D array of numbers specifing intensity.  Let's practice by plotting a 2D array of random numbers. Examples below are based on those [here](http://www.ds100.org/sp18/assets/lectures/lec07/07-intro-matplotlib.html). 

```python
a = np.random.rand(5,10)
a
```

```python
# In the command below, we are plotting the values in the array
# interpolation alows for different kinds of smothing between adjacent values
# cm.Blues means use the "Blues" colormap
plt.imshow(a, interpolation='none', cmap=plt.cm.Blues)
plt.figure() # create a new figure plot, otherwise the next command will plot on top of what was plotted previously

plt.imshow(a, interpolation='bilinear', cmap=plt.cm.BuPu)
plt.figure()
plt.imshow(a, interpolation='bicubic', cmap=plt.cm.Blues)
plt.figure()
plt.imshow(a, interpolation='nearest', cmap=plt.cm.Blues);
```

Now we will read in a color image.  

```python
img = plt.imread('https://www.andrews.edu/~tzs/DataScience/tulips.png')
img.shape
```

Notice that this color image is a 320 by 240 by 3 array.  320 by 240 is the image size in pixels. There are 3 separate 320 by 240 arrays that hold intensity values for red, green and blue (RGB).

```python
plt.imshow(img);
```

### Subplots
Plot the r, g, b channels of the image. If we want to directly compare the intensity of the color data in each channel, it's visually clearest if we do so by showing all individual channels as grayscale.

With the call clim=(0,1), we ensure that the visual range of the grayscale colormap spans the whole (0, 1) range for each channel, so that visual comparisons across make sense. Otherwise matplotlib would use the entire visual range adapted to each channel's data, making that comparison misleading.

```python
fig, ax = plt.subplots(1,4, figsize=(10,6))  # Make an array of subplots with 1 row and 4 columns
ax[0].imshow(img[:,:,0], cmap=plt.cm.Greys_r, clim=(0, 1)) # Plot red values in the first subplot
ax[1].imshow(img[:,:,1], cmap=plt.cm.Greys_r, clim=(0, 1))
ax[2].imshow(img[:,:,2], cmap=plt.cm.Greys_r, clim=(0, 1))
ax[3].imshow(img);
for a in ax:
    a.set_xticklabels([]); a.set_xticks([]) # Remove plot ticks and labels
    a.set_yticklabels([]); a.set_yticks([])
```

## 3D Plots

There is a very handy reference for all sorts of 3D plots using matplotlib [here](https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html).  One of those examples is reproduced below.

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.viridis,
        linewidth=0, antialiased=False)
ax.set_zlim3d(-1.01, 1.01);
```

## Scatterplots

Scatterplots are one of the simplest ways of investigating relationships between columns of data. Examples below are based on those [here](http://www.ds100.org/sp18/assets/lectures/lec07/07-intro-matplotlib.html). 

```python
import pandas as pd

cars = pd.read_csv('https://raw.githubusercontent.com/Brunel-Visualization/Brunel/master/R/examples/data/cars.csv')
cars.head()
```

```python
x, y = 'horsepower', 'mpg'

plt.scatter(x=x, y=y, data=cars);
```

```python
fig, ax = plt.subplots()
for name, df in cars.groupby('origin'):
    ax.scatter(x=x, y=y, data=df, label=name)

ax.set_xlabel(x)
ax.set_ylabel(y)
ax.legend();
```

```python
# Task 3: Make scatterplots of acceleration vs horsepower and horsepower vs cylinders in separate figures.  
# y vs x is the convention. That means that for acceleration vs horsepower, acceleration goes on the y
# axis and horsepower goes on the x axis.  Label the axes and give each plot a title.
x, y = 'horsepower', 'acceleration'
plt.scatter(x=x, y=y, data=cars);
plt.title('acceleration vs horsepower')
plt.xlabel('horsepower')
plt.ylabel('acceleration')
plt.show()

```

```python
x, y = 'cylinders', 'horsepower'
plt.scatter(x=x, y=y, data=cars);
plt.title('horsepower vs cylinders')
plt.xlabel('cylinders')
plt.ylabel('horsepower')

plt.show()
```

### Seaborn

Created by a Stanford graduate student, [Mike Waskom](https://seaborn.pydata.org/).

*Python visualization library based on matplotlib. It provides a high-level interface for drawing attractive statistical graphics.*

"...built on top of matplotlib and tightly integrated with the PyData stack, including support for `numpy` and `pandas` data structures and statistical routines from `scipy` and `statsmodels`."

Some of the features that seaborn offers are

* Several built-in themes that improve on the default matplotlib aesthetics
* Tools for choosing color palettes to make beautiful plots that reveal patterns in your data
* Functions for visualizing univariate and bivariate distributions or for comparing them between subsets of data
* Tools that fit and visualize linear regression models for different kinds of independent and dependent variables
* Functions that visualize matrices of data and use clustering algorithms to discover structure in those matrices
* A function to plot statistical timeseries data with flexible estimation and representation of uncertainty around the estimate
* High-level abstractions for structuring grids of plots that let you easily build complex visualizations

"The plotting functions operate on dataframes and arrays containing a whole dataset and internally perform the necessary aggregation and statistical model-fitting to produce informative plots. If matplotlib tries to make easy things easy and hard things possible, seaborn tries to make a well-defined set of hard things easy too."

```python
import seaborn as sns

x, y = 'cylinders', 'mpg'

sns.lmplot(x=x, y=y, data=cars, hue='origin', fit_reg=False)
ax = plt.gca()
ax.set_ylabel('Miles Per Gallon');
```

```python
# Add linear fits, which may or may not be meaningful in this case

sns.lmplot(x=x, y=y, data=cars, hue='origin');
```

```python
# Scatterplot everything vs everything
sns.pairplot(cars, hue="origin");
```

## Boxplots

We can use seaborn to make statistical boxplots. The example below is based on the one [here](https://seaborn.pydata.org/examples/horizontal_boxplot.html)

```python
f, ax = plt.subplots(figsize=(7, 6))

# Plot the mpg with vertical boxes
sns.boxplot(x="cylinders", y="mpg", data=cars, palette="vlag")

# Add in points to show each data point
sns.swarmplot(x="cylinders", y="mpg", data=cars,
              size=2, color=".3", linewidth=0)

# Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="Miles Per Gallon");
```

```python
# Task 4: Create a box plot of horsepower vs cylinders (horsepower on y, cylinders on x)
f, ax = plt.subplots(figsize=(7, 6))

sns.boxplot(x="cylinders", y="horsepower", data=cars, palette="vlag")

sns.swarmplot(x="cylinders", y="horsepower", data=cars,
              size=2, color=".3", linewidth=0)

ax.xaxis.grid(True)
ax.set(ylabel="horsepower");
```

## Pie Charts

The example below shows how to create a pie chart for some simple data. The example is from [here](https://pythonspot.com/matplotlib-pie-chart/) while documentation on the `pie` function is [here](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pie.html).

```python
# Data to plot
labels = 'mpg', 'cylinders', 'engine', 'horsepower','horsepower' 
sizes = [215, 130, 245, 210] # The fractional area of each wedge is given by sizes[i]/sum(sizes)
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```

```python
# Task 5: Make a pie chart showing the percentage of cars in the dataset with
# each different number of cylinders.  The calculation of the number of cars with
# each cylinder type has already been done to get you started.

from collections import Counter

carsdict = Counter(cars.cylinders)
ncyl = carsdict.keys()
ncars = carsdict.values()

labels = ncyl
sizes = ncars
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'blue']
explode = (0.1, 0, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=ncyl, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```
